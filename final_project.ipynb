```python
import datetime
def season_of_year(month, day):
    if month <= 2 or (month == 3 and day <= 21):
        return 1
    elif month <= 5 or (month == 6 and day <= 21):
        return 2
    elif month <= 8 or (month == 9 and day <= 21):
        return 3
    elif month <= 11 or (month == 12 and day <= 21):
        return 4
    else:
        return 1

data["Month"] = pd.to_datetime(data["Datum"]).dt.month
data["Day"] = pd.to_datetime(data["Datum"]).dt.day
data["Season"] = data.apply(lambda x: season_of_year(x['Month'], x["Day"]), axis=1)
data.head(5)
```

```python
data_groupped = data[["PM10", "PM2_5", "NO_", "NO2", "Nox", "O3", "SO2", "CO", "X" ,"Y", "Season"]].groupby(["X", "Y", "Season"]).agg('mean')
data_groupped.head(6)
```

```python
# Precomputing all values, this cell can take some time. If necessesary lower variable times to something smaller.

def count_value(x, y, season):
    filtered_season = data_groupped[data_groupped.index.map(lambda x: x[2] == season)]

    sum_dist = 0
    sum_value = {}
    for col in filtered_season:
        sum_value[col] = 0
        
    for i in filtered_season.index:
        dist2 = (i[0] - x) ** 2 + (i[1] - y ) ** 2
        for col in data_groupped:
            sum_value[col] += filtered_season[col][i] / dist2
        sum_dist += 1 / dist2

    for col in filtered_season:
        sum_value[col] /= sum_dist
        
    return sum_value


def calculate_data(times=50):
    left_upper_X = 49.2566
    left_upper_Y = 16.5253
    right_bottom_X = 49.1563
    right_bottom_Y = 16.6840
    
    new_data = pd.DataFrame()
    new_data["x"] = 0
    new_data["y"] = 0
    new_data["season"] = 0
    for col in data_groupped:
        new_data[col] = 0

    index_i = 0
    for x in [left_upper_Y + i * (right_bottom_Y - left_upper_Y) / times for i in range(times)]:
        for y in [left_upper_X + i * (right_bottom_X - left_upper_X) / times for i in range(times)]:
            for season in range(1, 5):
                row = {"x": x, "y": y, "season": season}
                row.update(count_value(y, x, season))
                new_data.loc[len(new_data)] = row

    return new_data

times=50
calculated_data = calculate_data(times=times)
calculated_data.head(10)
```

```python
from dash import Dash, html, dcc, Input, Output, dash_table
import plotly.graph_objects as go

app = Dash(__name__)

#*******APP LAYOUT**************

app.layout = html.Div(
    style={'backgroundColor':'#200010',
        'height': '100%',
        'color': 'white',
        'margin': 0,
        'padding': '18px',
         # 'background-image': 'url("/assets/brno_map.jpg")',
    }, 
    
    children=[
        #****************************HEADER*************************************
        html.H1(children='Air Pollution in Brno'),

        dcc.Markdown('### Visualization of harmful substances in the air measured by a mobile methodologist station'),
    
        html.Hr(),
    
        #****************************Controls*************************************
        html.Div(children=[
            'Fixed color range:', 
            dcc.RadioItems(['Fix', 'Auto'],
                       'Fix',
                       id='fix_scale',
                       inline=True)],
            style={'width': '100%', 'display': 'inline-flex'}
        ), 
        html.Hr(),

        html.Div(children=[
            'Substance:', 
            dcc.RadioItems(['PM10', 'PM2_5', 'NO_', 'NO2', 'Nox', 'O3', 'SO2', 'CO'], #options
                           'PM10', #
                           id='content',
                           inline=True)],
            style={'width': '100%', 'display': 'inline-flex'}
        ),
        html.Hr(),

        html.Div(children=[
            'Season of the Year:', 
            dcc.RadioItems(['Spring', 'Summer', 'Autumn', 'Winter'],
                       'Spring',
                       id='season',
                       inline=True)],
            style={'width': '100%',
                   'display': 'inline-flex', 
                  }
        ),
        html.Hr(),
        
#****************************Plots*************************************
        html.Div(
            dcc.Graph(id='map'), 
            style={
                   'width': '50%',
                   'height': '700px',
                   'display': 'inline-block'}
        ),

        html.Div(
            [dcc.Graph(id='substances')], 
            style={'width': '50%', 'height': '700px', 'display': 'inline-block'}),
        
        html.Div(
            dcc.Graph(id='timeline'), 
            style={
                   'width': '50%',
                   'display': 'inline-block'}
        ),
    ],
 )


def change_color(plotly_color, opacity=1.0):
    chgd_plotly_color=[]
    for color in plotly_color:
        color_string = color.replace('rgb', 'rgba')
        color_string = color_string.replace(')',f', {str(opacity)})')
        chgd_plotly_color.append(color_string)
    return chgd_plotly_color


def get_map(content, season, fix):
    left_upper_X = 49.2566
    left_upper_Y = 16.5253
    right_bottom_X = 49.1563
    right_bottom_Y = 16.6840
    season_dict = {"Winter":1, "Spring":2, "Summer":3, "Autumn": 4}

    range_max = max(calculated_data[content]) if fix=='Fix' else max(calculated_data[calculated_data["season"] == season_dict[season]][content])
    fig = px.density_heatmap(calculated_data[calculated_data["season"] == season_dict[season]],
                             x="x", y="y", z=content, histfunc="avg",
                             title=f"Content of {content} in Brno (µg/m³)",
                             nbinsx=times,
                             nbinsy=times,
                             range_color=(0, range_max),
                             height=700,
                            color_continuous_scale=change_color(px.colors.sequential.Reds, opacity=0.6),
                            )
        
    stations = list(set(map(lambda x: (x[0], x[1]), list(data_groupped.index))))
    stations = list(filter(lambda x: right_bottom_X <= x[0] <= left_upper_X and left_upper_Y <= x[1] <= right_bottom_Y, stations))  
    x0 = list(map(lambda x: x[1], stations)) 
    y0 = list(map(lambda x: x[0], stations))
    fig.add_trace(go.Scatter(
    x=x0,
    y=y0,
    mode='markers',
    text=["Methodogolist station" for k in x0],
    marker=dict(
        symbol='circle',
        opacity=0.7,
        color='black',
        size=8,
        line=dict(width=1),)
    ))
    
    fig.add_layout_image(
        dict(
            source="assets/brno_map_resize.jpg",
            xref="x",
            yref="y",
            x = left_upper_Y,
            y = left_upper_X,
            sizex=right_bottom_Y - left_upper_Y,
            sizey=right_bottom_X - left_upper_X,
            sizing="stretch",
            opacity=1,
            layer="below"))

    fig.update_layout(yaxis_title=None)
    fig.update_layout(xaxis_title=None)
    fig.update_xaxes(visible=False)
    fig.update_yaxes(visible=False)    
    return fig


def get_point_graph(x, y, season):
    season_dict = {"Winter":1, "Spring":2, "Summer":3, "Autumn": 4}
    d = calculated_data[calculated_data["season"] == season_dict[season]].copy()
    d["distance"] = (d["x"] - x) ** 2 + (d["y"] - y) ** 2
    
    closest = dict(d[d["distance"] == d["distance"].min()].iloc[0])
    true_x = closest["x"]
    true_y = closest["y"]
    closest.pop("x")
    closest.pop("y")
    closest.pop("season")
    closest.pop("distance")
    closest["CO * 10^-2"] = closest["CO"] / 1000
    closest["O3 * 10^-1"] = closest["O3"] / 10
    closest.pop("O3")
    closest.pop("CO")

    clean_air = {"PM10": 15,
                 "PM2_5": 5,
                 "NO_": 30,
                 "NO2": 10,
                 "Nox": 30,
                 "SO2": 40,
                 "CO * 10^-2": 40,
                 "O3 * 10^-1":10,
                }

    names = list(closest) + list(clean_air)
    values = [closest[k] for k in closest] + [clean_air[k] for k in clean_air]
    fig = px.bar(x=names, y=values, color=[f"lokality: {true_x:.4f}E {true_y:.4f}N" if i < 8 else "clean air" for i in range(16)], barmode="group", 
                 title=f"Content of substances at location {true_x:.4f}E {true_y:.4f}N",
                 labels={
                     "x": "Substance",
                     "y": "Content in the air (µg/m³)",
                     "color": "Type of data"
                 },
                 height=700,
                )

    fig.update_traces(marker_color=px.colors.sequential.Reds[2], selector={"name": "clean air"})
    fig.update_traces(marker_color=px.colors.sequential.Reds[-1], selector={"name": f"lokality: {true_x:.4f}E {true_y:.4f}N"})
    return fig


def get_timeline(x, y, season, content):
    season_dict = {"Winter":1, "Spring":2, "Summer":3, "Autumn": 4}
    d = data[data["Season"] == season_dict[season]].copy()
    d["distance"] = (d["X"] - y) ** 2 + (d["Y"] - x) ** 2
    closest = d[d["distance"] == d["distance"].min()].copy()
    closest["dayMonth"] = closest.apply(lambda row: f'{row["Day"]}. {row["Month"]}.', axis=1)

    if season == "Winter":
        closest["Month"] = closest["Month"].apply(lambda x: -1 if x == 12 else x)
    closest = closest.sort_values(by=["Month", "Day"])
    fig = px.line(closest, x="dayMonth", y=content, title=f'Values in time for methodologist station at {closest["Y"].iloc[0]}E {closest["X"].iloc[0]}N')

    fig.update_traces(line_color=px.colors.sequential.Reds[-1])
    return fig
    

@app.callback(
    Output('map', 'figure'),
    Input('content', 'value'),
    Input('season', 'value'),
    Input('fix_scale', 'value')
)
def update_map(content, season, fix):
    fig = get_map(content, season, fix)
    return fig


@app.callback(
    Output('substances', 'figure'),
    Input('map', 'clickData'),
    Input('season', 'value'),
)
def update_substances(clickData, season):
    x = 16.6086 #Brno Orloj coordinates
    y = 49.1948
    if clickData is not None:
        x = clickData["points"][0]["x"]
        y = clickData["points"][0]["y"]
    
    fig = get_point_graph(x, y, season)
    return fig


@app.callback(
    Output('timeline', 'figure'),
    Input('map', 'clickData'),
    Input('season', 'value'),
    Input('content', 'value'),
)
def update_timeline(clickData, season, content):
    x = 16.6086 #Brno Orloj coordinates
    y = 49.1948
    if clickData is not None:
        x = clickData["points"][0]["x"]
        y = clickData["points"][0]["y"]
    
    fig = get_timeline(x, y, season, content)
    return fig


#********RUNNING THE APP*************************************************
if __name__ == '__main__':
    app.run_server(jupyter_mode="external", debug=True, assets_folder='assets', assets_url_path='/assets/') # inline/tab/external
```
